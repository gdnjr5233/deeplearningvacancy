```python
import pandas as pd
from lxml import etree
```

## Вакансии

```python
vac = pd.read_csv('hhparser_vacancy.csv')
vac.head(5)
```

```python
description = vac["description"].head(5)[0]
description
```

```python
def removals(description):
    parser = etree.HTMLParser()
    tree = etree.fromstring(description, parser)
    return etree.tostring(tree, encoding='unicode', method='text')
```

```python
removals(description)
```

```python
vac["description"] = vac["description"].apply(lambda text: removals(description))
vac
```

```python
prepared_path= "/content/sample_data/vac_prepared.csv"
vac.to_csv(prepared_path, index=False)
```

```python
vac = pd.read_csv(prepared_path)
vac
```
