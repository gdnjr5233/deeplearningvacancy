```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assume df_jobs is your job dataset, and df_resumes is your resume dataset

# Extract relevant features for job descriptions and resumes
job_descriptions = df_jobs['description'].fillna('')
resume_experience = df_resumes['Опыт работы'].fillna('')

# Replace NaN values with an empty string to avoid issues with SpaCy
job_descriptions = job_descriptions.replace({pd.NA: ''})
resume_experience = resume_experience.replace({pd.NA: ''})

# Combine descriptions for TF-IDF vectorization
all_descriptions = job_descriptions.append(resume_experience)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_descriptions)

# Calculate cosine similarity between job descriptions and resume experience
experience_similarity = cosine_similarity(tfidf_matrix[-len(resume_experience):], tfidf_matrix[:-len(resume_experience)])

# Find the best-matching jobs for each resume based on experience
best_matching_jobs_indices = experience_similarity.argmax(axis=1)

# Add a column to the resume dataset with the corresponding best-matching job indices
df_resumes['best_matching_job_index'] = best_matching_jobs_indices

# Access the corresponding job IDs using the indices
df_resumes['best_matching_job_id'] = df_jobs.iloc[best_matching_jobs_indices].index

```

```python
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
from tqdm import tqdm

# Download NLTK resources (stopwords)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Define a function to clean up text
def clean_text(text):
    # Remove HTML tags
    soup = BeautifulSoup(text, 'html.parser')
    cleaned_text = soup.get_text(separator=' ')

    # Remove non-alphanumeric characters
    cleaned_text = re.sub(r'[^a-zA-Zа-яА-Я0-9\s]', '', cleaned_text)

    # Convert to lowercase
    cleaned_text = cleaned_text.lower()

    # Tokenize the text
    tokens = nltk.word_tokenize(cleaned_text,language='russian')

    # Remove stopwords
    stop_words = set(stopwords.words('russian'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Join the filtered tokens back into text
    cleaned_text = ' '.join(filtered_tokens)

    return cleaned_text

# Apply the clean_text function to the 'Опыт работы' column in df_resumes
df_resumes['cleaned_experience'] = df_resumes['Опыт работы'].apply(clean_text)
df_jobs['cleaned_description'] = df_jobs['description'].apply(clean_text)
# Print the cleaned DataFrame
print(df_resumes[['Опыт работы', 'cleaned_experience']])
print(df_jobs[['description', 'cleaned_description']])
```

```python
import spacy
import numpy as np
import time
from spacy.lang.ru.examples import sentences

# Load Spacy model
nlp = spacy.load("ru_core_news_sm")

def calculate_batch_semantic_similarity(texts1, texts2):
    # Process texts in batches
    docs1 = list(nlp.pipe(texts1, disable=["tagger", "parser", "ner"]))
    docs2 = list(nlp.pipe(texts2, disable=["tagger", "parser", "ner"]))

    # Calculate similarity scores
    similarities = [doc1.similarity(doc2) for doc1, doc2 in zip(docs1, docs2)]
    return np.array(similarities)

# Batch processing of semantic similarity with time tracking
batch_size = 100  # Adjust the batch size based on your system's capabilities
batches = range(0, len(df_jobs_test), batch_size)

start_time = time.time()

# Initialize an empty list to store the best matching job IDs
best_matching_job_ids = []

# Batch processing of semantic similarity with time tracking
batch_size = 100  # Adjust the batch size based on your system's capabilities
batches = range(0, len(df_jobs_test), batch_size)

start_time = time.time()

for i, batch_start in enumerate(batches):
    batch_end = min(batch_start + batch_size, len(df_jobs_test))

    # Extract relevant job and resume texts for the batch
    job_descriptions_batch = df_jobs_test.iloc[batch_start:batch_end]['cleaned_description'].tolist()
    resume_experiences_batch = df_resumes_test['cleaned_experience'].tolist()

    # Calculate semantic similarity for the batch
    batch_start_time = time.time()
    batch_similarity = calculate_batch_semantic_similarity(job_descriptions_batch, resume_experiences_batch)
    batch_end_time = time.time()

    # Get the index of the job with the maximum similarity in the batch
    max_similarity_index = np.argmax(batch_similarity)

    # Get the corresponding job ID
    best_matching_job_id = df_jobs_test.iloc[batch_start + max_similarity_index]['id']

    # Append the job ID to the list
    best_matching_job_ids.extend([best_matching_job_id] * (batch_end - batch_start))

    # Print progress information
    print(f"Processed batch {i + 1}/{len(batches)} in {batch_end_time - batch_start_time:.2f} seconds")

# Assign the list of best matching job IDs to the DataFrame
df_resumes_test['best_matching_job_id_semantic'] = best_matching_job_ids

total_time = time.time() - start_time
print(f"Total time taken: {total_time:.2f} seconds")


```

```python
import spacy
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


from gensim.models import Word2Vec
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Assuming you have preprocessed columns: 'preprocessed_experience', 'preprocessed_previous_job', 'preprocessed_description', 'preprocessed_job_tittle'

# Combine the preprocessed text columns into a new column for training the Word2Vec model
df_resumes_test['combined_text'] = df_resumes_test.apply(lambda row: ' '.join(row['preprocessed_experience'] + row['preprocessed_previous_job']), axis=1)
df_jobs_test['combined_text'] = df_jobs_test.apply(lambda row: ' '.join(row['preprocessed_description'] + row['preprocessed_job_tittle']), axis=1)


# Train a Word2Vec model on the combined text from both resumes and job descriptions
documents = df_resumes_test['combined_text'].append(df_jobs_test['combined_text'], ignore_index=True).tolist()
tokenized_documents = [doc.split() for doc in documents]

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_documents, vector_size=100, window=5, min_count=1, workers=4)

# Create a TF-IDF vectorizer for transforming text to numerical vectors
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Transform the resumes and jobs text to TF-IDF vectors
resumes_tfidf = tfidf_matrix[:len(df_resumes_test)]
jobs_tfidf = tfidf_matrix[len(df_resumes_test):]

# Calculate cosine similarity between resumes and job descriptions using TF-IDF
tfidf_similarity = cosine_similarity(resumes_tfidf, jobs_tfidf)

# Calculate cosine similarity using Word2Vec embeddings
def calculate_word2vec_similarity(text1, text2):
    tokens1 = text1.split()
    tokens2 = text2.split()
    vector1 = np.mean([word2vec_model.wv[token] for token in tokens1 if token in word2vec_model.wv], axis=0)
    vector2 = np.mean([word2vec_model.wv[token] for token in tokens2 if token in word2vec_model.wv], axis=0)
    return cosine_similarity([vector1], [vector2])[0][0]

word2vec_similarity = df_resumes_test.apply(
    lambda resume: df_jobs_test.apply(
        lambda job: calculate_word2vec_similarity(resume['combined_text'], job['combined_text']),
        axis=1
    ),
    axis=1
)

# Load the spaCy model for the Russian language
nlp = spacy.load("ru_core_news_sm")

# Function to calculate semantic similarity using spaCy
def calculate_spacy_similarity(text1, text2):
    doc1 = nlp(text1)
    doc2 = nlp(text2)
    return doc1.similarity(doc2)


# Calculate Spacy similarity
spacy_similarity = df_resumes_test.apply(
    lambda resume: df_jobs_test.apply(
        lambda job: calculate_spacy_similarity(resume['combined_text'], job['combined_text']),
        axis=1
    ),
    axis=1
)

# Combine the similarity scores (adjust the weights based on your preference)
combined_similarity = 0.5 * experience_similarity + 0.3 * word2vec_similarity + 0.2 * spacy_similarity

# Find the best-matching jobs for each resume based on combined similarity
best_matching_jobs_indices = combined_similarity.idxmax(axis=1)

# Add a column to the resume dataset with the corresponding best-matching job IDs
df_resumes['best_matching_job_id'] = df_jobs.iloc[best_matching_jobs_indices]['id'].values
```
